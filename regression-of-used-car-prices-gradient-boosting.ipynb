# Cleaning Data

```python
%%time

train_data_filled = train.copy()
test_data_filled = test.copy()
```

```python
%%time

# For categorical/text columns, fill missing values with 'Unknown'
train_data_filled['fuel_type'].fillna('Unknown', inplace=True)
train_data_filled['accident'].fillna('Unknown', inplace=True)
train_data_filled['clean_title'].fillna('Unknown', inplace=True)

test_data_filled['fuel_type'].fillna('Unknown', inplace=True)
test_data_filled['accident'].fillna('Unknown', inplace=True)
test_data_filled['clean_title'].fillna('Unknown', inplace=True)
```

```python
%%time

# For numeric columns, fill missing values with 0
train_data_filled.fillna(0, inplace=True)
test_data_filled.fillna(0, inplace=True)
```

# Feature Engineering For Train Data

```python
%%time

train_data_fe = train_data_filled.copy()
```

* Create a feature for car age based on model_year

```python
%%time

current_year = 2024  # assuming current year
train_data_fe['car_age'] = current_year - train_data_fe['model_year']

train_data_fe['car_age']
```

* Calculate mileage per year

```python
%%time

# To avoid division by zero in case car_age is 0, replace those values with 1

train_data_fe['car_age'] = train_data_fe['car_age'].replace(0, 1)
train_data_fe['mileage_per_year'] = train_data_fe['milage'] / train_data_fe['car_age']

train_data_fe['mileage_per_year']
```

# Feature Engineering For Test Data

```python
%%time

test_data_fe = test_data_filled.copy()
```

* Create a feature for car age based on model_year

```python
%%time

current_year = 2024  # assuming current year
test_data_fe['car_age'] = current_year - test_data_fe['model_year']

test_data_fe['car_age']
```

* Calculate mileage per year

```python
%%time

test_data_fe['car_age'] = test_data_fe['car_age'].replace(0, 1)
test_data_fe['mileage_per_year'] = test_data_fe['milage'] / test_data_fe['car_age']

test_data_fe['mileage_per_year']
```

# Train Using h2o automl

```python
%%time

import h2o
from h2o.automl import H2OAutoML

# Start the H2O cluster
h2o.init()
```

```python
%%time

# Convert Pandas DataFrames to H2O frames
train_h2o = h2o.H2OFrame(train_data_filled)
test_h2o = h2o.H2OFrame(test_data_filled)
```

* Set price column as target

```python
%%time

# Define target and predictors
target = 'price'
predictors = train_h2o.columns
```

```python
%%time

predictors.remove('id') 
```

```python
%%time

# Split the training data into train and validation sets
train, valid = train_h2o.split_frame(ratios=[0.8], seed=42)
```

```python
%%time

# Initialize H2O AutoML
aml = H2OAutoML(max_models=20, seed=42, exclude_algos=["DeepLearning"], stopping_metric='RMSE', max_runtime_secs=1000)

# Train the model
aml.train(x=predictors, y=target, training_frame=train, validation_frame=valid)
```

```python
%%time 

# View the leaderboard
lb = aml.leaderboard
print(lb)
```

```python
%%time

# Make predictions on the test set
preds = aml.predict(test_h2o)
```

```python
%%time

# Convert H2OFrame predictions to a pandas DataFrame
predictions_df = preds.as_data_frame()
```

```python
%%time

# Prepare the submission file
submission = pd.DataFrame({'id': test_data_filled['id'], 'price': predictions_df['predict']})
```

```python
%%time

# Create submission file
submission.to_csv('submission.csv', index=False)
```
